Write a Python program that:
* Reads the file `notebook/01-publications-venues.json`
* For each publication that contains a "publication venue", extracts: the publication title and the name of the publication venue
* Writes this information to a new CSV file called with the columns "title" and "venue_name"

```python
print("-------------------- SOLUTION --------------------")

from json import load

pub_venues = []
with open("../notebook/01-publications-venues.json", "r", encoding="utf-8") as f:
    json_pub = load(f)
    for p in json_pub:
        if 'publication venue' in p:
            pub_venues.append(
                {
                    "title": p["title"],
                    "venue_name": p["publication venue"]["name"]
                }
            )
        

from csv import DictWriter
with open("ex-15-2-01-publications-venue-names.csv", "w", encoding="utf-8") as f:
    
    header = ["title","venue_name"]
    dwr = DictWriter(f, header)
    dwr.writeheader()
    dwr.writerows(pub_venues)
```
