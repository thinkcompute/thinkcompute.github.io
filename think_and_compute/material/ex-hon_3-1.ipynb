Create a pandas DataFrame by combining two Series: one containing five countries and the other their corresponding capitals. Then sort the DataFrame alphabetically by country name.

```python
import pandas as pd

# Create two Series
countries = pd.Series([
    "Italy",
    "France",
    "Germany",
    "Spain",
    "Portugal"
])

capitals = pd.Series([
    "Rome",
    "Paris",
    "Berlin",
    "Madrid",
    "Lisbon"
])

# Combine Series into a DataFrame
df = pd.DataFrame({
    "Country": countries,
    "Capital": capitals
})

# Sort DataFrame alphabetically by country
df_sorted = df.sort_values(by="Country").reset_index(drop=True)

df_sorted
```
