Using the rdflib library in Python, create RDF triples for the following resources:
* `https://hon-lecture/6/solution/1`
* `https://hon-lecture/6/solution/2`
* `https://hon-lecture/6/solution/3`

For each resource, define three RDF triples:
1. A triple stating that the resource is of type `solution`.
2. A triple specifying the name of the resource using the literal format `"solution-N"` (for example, `"solution-1"`).
3. A triple which uses the relation `"solutionOf"` to indicate that a `solution-N` the actual solution of `exercise-N` (copy the solution of `ex-hon_6-1` into your code )

```python
from rdflib import Graph, URIRef, Literal
from rdflib.namespace import RDF

# Create graph
g = Graph()

## SOLUTION OF 6-1
## ---------------
# Define class and property
exercise = URIRef("https://hon-lecture/Class/Exercise")
name = URIRef("https://schema.org/name")
solution = URIRef("https://hon-lecture/Class/Solution")
solutionOf = URIRef("https://hon-lecture/solutionOf")

# Create triples
for i in range(1, 4):

    # SOLUTION of 6-1
    # ---------------
    ex_resource = URIRef(f"https://hon-lecture/6/ex/{i}")
    g.add((ex_resource, RDF.type, exercise))
    g.add( ( ex_resource, name, Literal("exercise-"+str(i)) ) )
    # ---------------

    resource = URIRef(f"https://hon-lecture/6/solution/{i}")
    # Triple 1: resource is of type solution
    g.add((resource, RDF.type, solution))

    # Triple 2: resource has a name "solution-N"
    g.add( ( resource, name, Literal("solution-"+str(i)) ) )

    # Triple 3: resource is a solution to the "exercise-N"
    g.add( ( resource, solutionOf, ex_resource ) )


# Print graph in Turtle format
print(g.serialize(format="turtle"))
```
