Make the class `Student` defined in the previous exercise a subclass of `Person`. Each `Person`entity must have a `name` and an `age`. Then define two students and a person and print all their ages.

```python
import random

class Person():
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def getName(self):
        return self.name

    def getAge(self):
        return self.age

class Student(Person):
    def __init__(self, name, age, faculty, university):
        self.faculty = faculty
        self.university = university
        self.id = faculty[0:2] + university[0:2] + str(random.randint(1, 10))
        
        super().__init__(name, age)
    
    def getID(self):
        return self.id

    def getFaculty(self):
        return self.faculty

    def getUNI(self):
        return self.university

st1 = Student("Edward",25,"Medicine","Harvard")
st2 = Student("James",24,"Computer Science","Bologna")
p1 = Person("Andrea",33)

print(st1.getAge())
print(st2.getAge())
print(p1.getAge())
```
