Using Pandas read the CSV file `notebook/01-venues.csv` and create a dataframe which includes the `name` and the `type` of the venue and a new column to use as internal identifier of such dataframe, the internal identifiers must have the following format: `V{NUMBER}`, e.g. `V01` or `V24`. Finally store the new dataframe inside an SQLITE DB.

```python
from sqlite3 import connect
from pandas import read_csv, Series

venues = read_csv("../notebook/01-venues.csv", 
                  keep_default_na=False,
                  dtype={
                      "id": "string",
                      "name": "string",
                      "type": "string"
                  })

venues_ids = venues[["name","type"]]

# Generate a list of internal identifiers for the venues
venue_internal_id = []
for idx, row in venues_ids.iterrows():
    venue_internal_id.append("V" + str(idx))

# Add the list of venues internal identifiers as a new column
# of the data frame via the class 'Series'
venues_ids.insert(0, "internalId", Series(venue_internal_id, dtype="string"))

# Show the new data frame on screen
with connect("my-venues.db") as con:
    venues_ids.to_sql("VenueId", con, if_exists="replace", index=False)
```
