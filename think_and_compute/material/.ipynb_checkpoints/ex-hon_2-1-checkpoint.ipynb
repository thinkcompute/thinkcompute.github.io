Create a Python class called `Student`. The class should have the attributes: `faculty (string)`, `university (string)`, and 
`id (string)`. The class should include GET methods for all these attributes.

When a `Student` object is created, it should automatically generate a student ID with this format:
* the first two letters of the faculty name;
* the first two letters of the university name;
* and a random integer between 1 and 10
(Example: if the `faculty="Medicine"`, `university="Oxford"`, then `id=MeOx3`)

Finaly, create two student objects:
* One with faculty "Medicine" and university "Harvard"
* One with faculty "Computer Science" and university "Bologna"
* Print the student IDs of both students.

```python
import random

class Student():
    def __init__(self, faculty, university):
        self.faculty = faculty
        self.university = university
        self.id = faculty[0:2] + university[0:2] + str(random.randint(1, 10))
    
    def getID(self):
        return self.id

    def getFaculty(self):
        return self.faculty

    def getUNI(self):
        return self.university

st1 = Student("Medicine","Harvard")
st2 = Student("Computer Science","Bologna")

print(st1.getID())
print(st2.getID())
```
