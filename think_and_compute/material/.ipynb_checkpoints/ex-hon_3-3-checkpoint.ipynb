Using pandas, load the file `notebook/01-publications.csv` into a DataFrame. Query the DataFrame to retrieve all publications with `publication_year == 2002`, and save the resulting DataFrame to a new CSV file named 01-publications-2002.csv. Save the new dataframe into a CSV with the inclusion of only `doi` and `publication year` columns.

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv("../notebook/01-publications.csv")

# Create a new row (example data — adjust columns as needed)
new_row = {
    "doi": "10.10/101010",
    "title": "The title of my publication",
    "publication year": "2020",
    "publication venue": "My venue",
    "type": "journal article",
    "issue": "",
    "volume": ""
}

# Add the new row to the DataFrame
df_2002 = df.query("`publication year`==2002")

# take just two columns
p_df_2002 = df_2002[["doi","publication year"]]

# Save the modified DataFrame to a new CSV file
p_df_2002.to_csv("01-publications-2020.csv", index=False)
```
