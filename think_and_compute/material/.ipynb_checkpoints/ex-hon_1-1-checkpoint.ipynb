Write a Python program that reads the file “01-publications.csv” and print the value from the title column for each row. Do not print the column header—only the title values.

```python

## SOLUTION 1
print("-------------------- SOLUTION 1 --------------------")

from csv import reader
with open("../notebook/01-publications.csv", "r", encoding="utf-8") as f:
    publications = reader(f)
    next(publications)
    for row in publications:
        print(row[1])


## SOLUTION 2
print("-------------------- SOLUTION 2 --------------------")

from csv import DictReader
with open("../notebook/01-publications.csv", "r", encoding="utf-8") as f:
    publications = DictReader(f)
    for p in publications:
        print(p["title"])

```
