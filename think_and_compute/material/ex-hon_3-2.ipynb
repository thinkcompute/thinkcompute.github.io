Using pandas, load the file `notebook/01-publications.csv` into a DataFrame, add one new row of data, and then save the updated DataFrame to a new CSV file named 01-publications-modified-with-pandas.csv. Don't save the index in the new CSV file.

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv("../notebook/01-publications.csv")

# Create a new row (example data — adjust columns as needed)
new_row = {
    "doi": "10.10/101010",
    "title": "The title of my publication",
    "publication year": "2020",
    "publication venue": "My venue",
    "type": "journal article",
    "issue": "",
    "volume": ""
}

# Add the new row to the DataFrame
len_of_df = len(df)
df.loc[len_of_df] = new_row

# Save the modified DataFrame to a new CSV file
df.to_csv("01-publications-modified-with-pandas.csv", index=False)
```
