Using pandas, load the CSV file `notebook/04-publications.csv` into a DataFrame. Filter the publications to include only those published between the years 2000 and 2010 (inclusive). Then compute the number of publications for each publication venue within this filtered dataset.

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv("../notebook/04-publications.csv")

# Filter publications between 2000 and 2010 (inclusive)
filtered_df = df.query('`publication year` >= 2000 and `publication year` <= 2010') 

# Count publications per venue
venue_counts = filtered_df["publication venue"].value_counts()

# Print results
print(venue_counts)
```
