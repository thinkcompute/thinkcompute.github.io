Using the rdflib library in Python, create RDF triples for the following resources:

* `https://hon-lecture/6/ex/1`
* `https://hon-lecture/6/ex/2`
* `https://hon-lecture/6/ex/3`

For each resource, define two RDF triples:
1. A triple stating that the resource is of type `exercise`.
2. A triple specifying the name of the resource using the literal format `"exercise-N"` (for example, `"exercise-1"`).

```python
from rdflib import Graph, URIRef, Literal
from rdflib.namespace import RDF

# Create graph
g = Graph()

# Define class and property
exercise = URIRef("https://hon-lecture/Class/Exercise")
name = URIRef("https://schema.org/name")

# Create triples
for i in range(1, 4):
    resource = URIRef(f"https://hon-lecture/6/ex/{i}")

    # Triple 1: resource is of type exercise
    g.add((resource, RDF.type, exercise))

    # Triple 2: resource has a name "exercise-N"
    g.add( ( resource, name, Literal("exercise-"+str(i)) ) )

# Print graph in Turtle format
print(g.serialize(format="turtle"))
```
