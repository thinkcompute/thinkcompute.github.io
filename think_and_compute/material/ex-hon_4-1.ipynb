Using pandas, read the CSV file `notebook/04-publications.csv` into a DataFrame. Filter the data to include only the publications whose `publication venue` is `"Materials Science Forum"`. Then determine the minimum and maximum `publication year` within this filtered DataFrame.

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv("../notebook/04-publications.csv")

# Filter rows where publication venue is "Materials Science Forum"
filtered_df = df[df["publication venue"] == "Materials Science Forum"]

# Get minimum and maximum publication year
min_year = filtered_df["publication year"].min()
max_year = filtered_df["publication year"].max()

# Print results
print("Minimum publication year:", min_year)
print("Maximum publication year:", max_year)
```
